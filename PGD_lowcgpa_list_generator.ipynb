```python
All_course_set= {'GST707','CIT701','CIT703','CIT753','CIT735','CIT722','CIT736','CIT756','CIT752','CIT799','CIT711','CIT759','CIT721','CIT734','CIT742'}
elective_course_set= {'CIT711','CIT759','CIT721','CIT734','CIT742'}
compulsory_course_set ={'GST707','CIT701','CIT703','CIT753','CIT735','CIT722','CIT736','CIT756','CIT752','CIT799'}
```

```python
def regularize_course_codes(course):
    new_course=[]
    for n in range(len(course)):
        srt = course[n][3]
        n_course = course[n]
        if srt == "6"  :
            n_course = course[n][:3] + '7' + course[n][4:]
            new_course.append(n_course)
        else:
            new_course.append(n_course)
        
    return new_course
def generate_passedCourse_list(graded_course):
    passed_course_list = []
    for i in range(len(graded_course)):
        course= graded_course[i][0:6]
        passed_course_list.append(course)
    return passed_course_list
```
